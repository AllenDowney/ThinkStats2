```python
grouped = df[['dayofweek', 'ppg']].groupby('dayofweek')
cdfs = grouped.aggregate(thinkstats2.Cdf)
thinkplot.Cdfs(cdfs.ppg)
```

```python
grouped = df[['days', 'ppg']].groupby('days')
ppg = grouped.aggregate(np.mean)
all_days = xrange(ppg.index.min(), ppg.index.max())
ppg = ppg.reindex(all_days)
thinkplot.Plot(ppg.index, ppg.ppg)
len(ppg.index), len(all_days)
```
